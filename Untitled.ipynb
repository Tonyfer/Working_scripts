```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
cwd = '/data2/HBNcore/CMI_HBN_Data/MRI/SI/Release_Derivatives/data_release_ExternalID/'
```

```python
files = os.listdir(cwd)
fi_list = []
for name in files:
    if 'sub' in name:
        fi_list.append(name)
```

```python
def get_v_a(s):
    if s in name_v_ant:
        return(volum_ant[volum_ant.columns[2]][name_v_ant.index(s)])
    else:
        return ''

def get_v_fs(s):
    if s in name_v_fs:
        return(volum_fs[volum_fs.columns[2]][name_v_fs.index(s)])
    else:
        return ''

def get_t_fs(s):
    if s in name_t_fs:
        return(thick_fs[thick_fs.columns[2]][name_t_fs.index(s)])
    else:
        return ''
    
def get_t_ant(s):
    if s in name_t_ant:
        return(thick_ant[thick_ant.columns[2]][name_t_ant.index(s)])
    else:
        return ''
```

```python
sub = cwd + 'sub-NDARAT299YRR'
```

```python
fi_list.remove('sub-NDARAT299YRR')
f_list = []
for path in fi_list:
    if (os.path.exists(cwd+path+'/mindboggle/')) and (os.path.exists(cwd+path+'/mindboggle/tables/left_cortical_surface/')) and (os.path.exists(cwd+path+'/mindboggle/tables/right_cortical_surface/')):
        f_list.append(path)
        
```

```python
thick_ant = pd.read_csv(sub+'/mindboggle/tables/thickinthehead_per_ants_cortex_label.csv')
thick_fs = pd.read_csv(sub+'/mindboggle/tables/thickinthehead_per_freesurfer_cortex_label.csv')
volum_fs = pd.read_csv(sub+'/mindboggle/tables/volume_per_freesurfer_label.csv')
volum_ant = pd.read_csv(sub+'/mindboggle/tables/volume_per_ants_label.csv')
lh = pd.read_csv(sub+'/mindboggle/tables/left_cortical_surface/label_shapes.csv')
rh = pd.read_csv(sub+'/mindboggle/tables/right_cortical_surface/label_shapes.csv')
```

```python
name_v_ant_template = [volum_ant['name'][i] for i in range(len(volum_ant['name']))]
name_v_fs_template = [volum_fs['name'][i] for i in range(len(volum_fs['name']))]
name_t_ant_template = [thick_ant['name'][i] for i in range(len(thick_ant['name']))]
name_t_fs_template = [thick_fs['name'][i] for i in range(len(thick_fs['name']))]
```

```python
name_v_ant = [volum_ant['name'][i] for i in range(len(volum_ant['name']))]
value_v_ant = [get_v_a(i) for i in name_v_ant_template]
name_v_ant = ['ant_v_'+volum_ant['name'][i] for i in range(len(volum_ant['name']))]
name_v_fs = [volum_fs['name'][i] for i in range(len(volum_fs['name']))]
value_v_fs = [get_v_fs(i) for i in name_v_fs_template]
name_v_fs = ['fs_v_'+volum_fs['name'][i] for i in range(len(volum_fs['name']))]
name_t_fs = [thick_fs['name'][i] for i in range(len(thick_fs['name']))]
value_t_fs = [get_t_fs(i) for i in name_t_fs_template]
name_t_fs = ['fs_t_'+thick_fs['name'][i] for i in range(len(thick_fs['name']))]
name_t_ant = [thick_ant['name'][i] for i in range(len(thick_ant['name']))]
value_t_ant = [get_t_ant(i) for i in name_t_ant_template]
name_t_ant = ['ant_t_'+thick_ant['name'][i] for i in range(len(thick_ant['name']))]
```

```python
name_s_a = ['area' + lh['name'][i] for i in range(len(lh['name']))]
va_l_a = [lh['area'][i] for i in range(len(lh['name']))]
va_r_a = [rh['area'][i] for i in range(len(rh['name']))]
name_s_tdm = ['tdm' + lh['name'][i] for i in range(len(lh['name']))]
va_l_t = [lh['travel depth: median'][i] for i in range(len(lh['name']))]
va_r_t = [rh['travel depth: median'][i] for i in range(len(rh['name']))]
val_a = [va_l_a[i]+va_r_a[i] for i in range(len(va_l_a))]
val_t = [va_l_t[i]+va_r_t[i] for i in range(len(va_l_t))]
```

```python
var = name_t_fs+name_t_ant+name_v_fs+name_v_ant+name_s_a+name_s_tdm
val = value_t_fs+value_t_ant+value_v_fs+value_v_ant+val_a+val_t
```

```python
df = pd.DataFrame(np.array(val).reshape(1,514), columns=var)
```

```python
i = 1
for sub in f_list:
    sub = cwd + sub
    thick_ant = pd.read_csv(sub+'/mindboggle/tables/thickinthehead_per_ants_cortex_label.csv')
    thick_fs = pd.read_csv(sub+'/mindboggle/tables/thickinthehead_per_freesurfer_cortex_label.csv')
    volum_fs = pd.read_csv(sub+'/mindboggle/tables/volume_per_freesurfer_label.csv')
    volum_ant = pd.read_csv(sub+'/mindboggle/tables/volume_per_ants_label.csv')
    lh = pd.read_csv(sub+'/mindboggle/tables/left_cortical_surface/label_shapes.csv')
    rh = pd.read_csv(sub+'/mindboggle/tables/right_cortical_surface/label_shapes.csv')
    name_v_ant = [volum_ant['name'][i] for i in range(len(volum_ant['name']))]
    value_v_ant = [get_v_a(i) for i in name_v_ant_template]
    #name_v_ant = ['ant_v_'+volum_ant['name'][i] for i in range(len(volum_ant['name']))]
    name_v_fs = [volum_fs['name'][i] for i in range(len(volum_fs['name']))]
    value_v_fs = [get_v_fs(i) for i in name_v_fs_template]
    #name_v_fs = ['fs_v_'+volum_fs['name'][i] for i in range(len(volum_fs['name']))]
    name_t_fs = [thick_fs['name'][i] for i in range(len(thick_fs['name']))]
    value_t_fs = [get_t_fs(i) for i in name_t_fs_template]
    #name_t_fs = ['fs_t_'+thick_fs['name'][i] for i in range(len(thick_fs['name']))]
    name_t_ant = [thick_ant['name'][i] for i in range(len(thick_ant['name']))]
    value_t_ant = [get_t_ant(i) for i in name_t_ant_template]
    #name_t_ant = ['ant_t_'+thick_ant['name'][i] for i in range(len(thick_ant['name']))]

    va_l_a = [lh['area'][i] for i in range(len(lh['name']))]
    va_r_a = [rh['area'][i] for i in range(len(rh['name']))]
    #name_s_tdm = ['tdm' + lh['name'][i] for i in range(len(lh['name']))]
    va_l_t = [lh['travel depth: median'][i] for i in range(len(lh['name']))]
    va_r_t = [rh['travel depth: median'][i] for i in range(len(rh['name']))]
    val_a = [va_l_a[i]+va_r_a[i] for i in range(len(va_l_a))]
    val_t = [va_l_t[i]+va_r_t[i] for i in range(len(va_l_t))]
    val = value_t_fs+value_t_ant+value_v_fs+value_v_ant+val_a+val_t
    df.loc[i] = val
    i+=1
    
    
    
    
    
```
